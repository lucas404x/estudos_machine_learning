### Tratando de dados faltantes

```python
for column in data.columns:
    print("Column: " + column)
    print(data.loc[pd.isnull(data[column])])
    print("***************")
```

##### A coluna "age" possuí valores faltantes. Como tratar?

```python
previsores = data.iloc[:,1:4].values # seleciona do "income" até o "loan"
atributo_classe = data['default'].values
```

### Para treinar meus algoritmos, SEMPRE pegar apenas os valores com "values"!

#### Base de dados -> data
#### previsores -> previsores
#### attr de classe -> atributo_classe

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='mean') # serve para completar os valores ausentes

imputer = imputer.fit(previsores[:, 0:3]) # serve para encaixar o objeto "imputer" na nossa base de dados

previsores[:, 0:3] = imputer.transform(previsores[:, 0:3]) # vai receber a transformação, ou seja, o modelo alterado com os valores substituidos
previsores
```
